```python
zp = data[['Work Location','Salary Range To']]

# (first+last)/2
zp.loc[:]['Salary Range To']=(data['Salary Range To']+data['Salary Range From'])/2
#zp.index =data['Work Location']

#group categories and add salaries
zp = zp.groupby('Work Location',sort=False).sum()

#change salary (category_sum/category_amount)
zp.loc[:]['Salary Range To']=zp['Salary Range To']/data['Work Location'].value_counts(sort=False)
zp[:20].plot(kind='bar')
```

```python
zp1 = data[['Job Category','Salary Range To']]

# (first+last)/2
zp1.loc[:]['Salary Range To']=(data['Salary Range To']+data['Salary Range From'])/2
#zp1.index =data['Job Category']

#group categories and add salaries
zp1 = zp1.groupby('Job Category',sort=False).sum()

#change salary (category_sum/category_amount)
zp1.loc[:]['Salary Range To']=zp1['Salary Range To']/data['Job Category'].value_counts(sort=False)
zp1[10:50].plot(kind='bar')
```
